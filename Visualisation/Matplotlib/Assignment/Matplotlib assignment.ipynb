## Q3

```python
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.figure(figsize=(10, 8))
```

```python
plt.subplot(2, 2, 1)
plt.plot(x1, y1, 'r-')
plt.title('Line 1')
plt.xlabel('x')
plt.ylabel('y')
plt.subplot(2, 2, 2)
plt.plot(x2, y2, 'g--')
plt.title('Line 2')
plt.xlabel('x')
plt.ylabel('y')
plt.subplot(2, 2, 3)
plt.plot(x3, y3, 'b-.')
plt.title('Line 3')
plt.xlabel('x')
plt.ylabel('y')
plt.subplot(2, 2, 4)
plt.plot(x4, y4, 'm:')
plt.title('Line 4')
plt.xlabel('x')
plt.ylabel('y')
plt.tight_layout()
plt.show()
```

## Q4


Barplot is used to represent descrete data (categorical data technically) it is shown by a collection of rectangle in a plor of cartesian plane quantifying the descrete data points for any givrn data points for a 2dimensional data as it peforms best in a 2d cartesian plane

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
# normal barplot
plt.xlabel('company')
plt.ylabel('profits')
plt.bar(company,profit)
```

```python
# vertical barplot
plt.xlabel('company')
plt.ylabel('profits')
plt.barh(company,profit)
```

## Q5

A box plot (or whisker plot) is a graphical representation of data that displays the distribution, variability, and summary statistics such as the median, quartiles, and potential outliers. It consists of a box that shows the interquartile range (IQR) and whiskers that extend to the minimum and maximum data points within 1.5 times the IQR. Points outside this range are considered outliers and are displayed as individual dots.

It is ised to view the data volume in a given dataset at aglance also


```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.figure(figsize=(8, 6))
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'], patch_artist=True)

# Add labels and title
plt.title('Box Plot of Two Distributions')
plt.ylabel('Values')
plt.xlabel('Boxes')

# Display the plot
plt.show()
```
