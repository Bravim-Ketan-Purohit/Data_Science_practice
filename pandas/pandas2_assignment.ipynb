## Q2

```python
def reindex(df,start=1,step=2):
  new_index = range(start,start + len(df) * step, step)
  df = df.copy()
  df.index = new_index
  return df
```

```python
data = {'A': [10, 20, 30],
        'B': [40, 50, 60],
        'C': [70, 80, 90]}
df1 = pd.DataFrame(data)
print("Original DataFrame:")
print(df1)
reindexed_df = reindex(df1)
print("\nRe-indexed DataFrame:")
print(reindexed_df)
```

## Q4

```python
data = {'Text': ['Hello world',
                 'This is a test',
                 'Pandas is powerful and flexible',
                 '',
                 None]}
df3 = pd.DataFrame(data)
df3
```

```python
len(df3)
```

```python
def word_counter(df):
  df['Word_Count'] = df['Text'].apply(lambda x: len(str(x).split()) if pd.notnull(x) else 0)
  return(df)
```

```python
print("Original")
print(df3)
df_ans = word_counter(df3)
print("\nDataFrame with Word_Count:")
print(df_ans)
```

## Q9

```python
data = {'Values': [10, 20, 30, 40, 50]}
df = pd.DataFrame(data)
```

```python
def stats(df):
  mean= df['Values'].mean()
  std = df['Values'].std()
  median = df['Values'].median()
  return({
      'mean': mean,
      'standard dev' : std,
      'median' : median
  })
```

```python
print(stats(df))
```

## Q10

```python
data = {
    'Sales': [100, 150, 200, 130, 120, 140, 160, 180, 220, 250],
    'Date': ['2024-12-01', '2024-12-02', '2024-12-03', '2024-12-04', '2024-12-05',
             '2024-12-06', '2024-12-07', '2024-12-08', '2024-12-09', '2024-12-10']
}
df = pd.DataFrame(data)
```

```python
df
```

```python
def avg(df):
  df['MovingAverage'] = df['Sales'].rolling(window=7, min_periods=1).mean()
  return(df)
```

## Q11

```python
data = {'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05']}
df = pd.DataFrame(data)
df
```

```python
def day_adder(df):
  df['Date'] = pd.to_datetime(df['Date'])
  df['Weekday'] = df['Date'].dt.day_name()
  return(df)
```

```python
day_adder(df)
```

```python
from datetime import date
def datepick(df):
    date1 = date(2023, 1, 1)
    date2 = date(2023, 1, 31)
    filtered_df = df[(df['Date'] > pd.Timestamp(date1)) & (df['Date'] < pd.Timestamp(date2))]

    return filtered_df
```
